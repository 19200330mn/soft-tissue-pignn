```python
from fenics import *
import matplotlib.pyplot as plt
import numpy as np
```

```python
FIGSIZE = (10,6)
save_dir = "results"
```

## 1. Load LV Mesh

```python
def load_mesh():
  
    mesh = Mesh("lvgmsh_mesh.xml")
    mesh.scale(0.1) # this will scale the mesh unit to be cm
    mesh.ufl_cell()
                
    mf = MeshFunction("size_t", mesh, "lvgmsh_fmaker.xml")

    
    # load fibers, sheet, cross_sheet data 
    fiber_element = VectorElement(family="Lagrange",
                                  cell=mesh.ufl_cell(),
                                  degree=1,
                                  quad_scheme="default")
    
    fiber_space = FunctionSpace(mesh, fiber_element)
    fibres      = Function(fiber_space, "fibre_dir.xml")

    numbering = {
        "BASE": 10,
        "ENDO": 30,
        "EPI": 40
    }
   
    return mesh, mf, numbering, fibres
```

```python
mesh, boundary_markers, numbering, fibres = load_mesh()
```

## 2. Extract mesh nodes and elements

```python
ref_coords = mesh.coordinates()
np.save(f'{save_dir}/reference-coords.npy', ref_coords)

elements   = mesh.cells()
np.save(f'{save_dir}/elements.npy', elements)
```

## 3. Extract mesh topology

```python
def get_mesh_topology(mesh_elements):
    
    def get_element_topology(element):
        # directed edge between each pair of nodes in the element (with no self loops)
        return [[i, j] for i in element for j in element if i != j]
    
    mesh_topology = []
    
    for element in mesh_elements:
        element_topology = get_element_topology(element)
        for edge in element_topology: mesh_topology.append(edge)
        
    # return unique edges in a numpy array
    return np.unique(np.array(mesh_topology), axis=0)
```

```python
real_node_topology = get_mesh_topology(elements)
np.save(f'{save_dir}/real-node-topology.npy', real_node_topology)
```

## 4. Extract fibre field

Both *node-wise* and *element-wise* - the node-wise fibre direction can be included in the feature vector assigned to each node, while the element-wise fibre direction is needed for the potential energy computations

First extract node-wise fibre field

```python
fibre_field = fibres.compute_vertex_values().reshape(ref_coords.T.shape).T
```

Then convert node-wise fibre field to element wise-fibre field. The fibre direction of each tetrahedral element is simply taken to be the mean of the fibres at each of its four vertices, normalised to unit length

```python
fibre_field_elementwise = fibre_field[elements].mean(1)

elementwise_norms = ((fibre_field_elementwise**2).sum(-1, keepdims=True))**.5

fibre_field_elementwise /= elementwise_norms 
np.save(f'{save_dir}/fibre-field-elementwise.npy', np.expand_dims(fibre_field_elementwise,-1))
```

## 5. Create indicator variable for nodes *not* on the Dirichlet boundary

For the LV mesh, this is simple - the Dirichlet points are those with $z$-coordinate value close to 0. So to find, we find those nodes with $z$-coord less than -0.10

```python
interior_points = ref_coords[:,2] < -0.1
np.save(f'{save_dir}/interior-points.npy', interior_points*1.)
```

## 6. Find indices of nodes on the endocardial surface

The endocardial surface is where the traction force is applied

Calling ``facets(mesh)`` returns an iterator that allows us to loop over all triangular facets/surfaces in the FE mesh (each facet is a 3-dimensional vector, the elements of which given the index of the corresponding node in the FE mesh).

To identify the facets on the endocardial surface, we loop over all facets (``for facet in facets(mesh):``), and find its boundary marker value (``boundary_markers[facet.index()]``). If this value matches the boundary marker of the endocardial surfaces (``numbering['ENDO']``), then we know it is on the endo surface, and we append this facet to a placeholder list (``node_numbering_list``).

```python
# place holder list
node_numbering_list = []

# loop over all triangular facets in the mesh
for facet in facets(mesh):
    # check if the marker of facet matches the endocardial marker
    if boundary_markers[facet.index()] == numbering['ENDO']:
        # if the markers match, then facet lies on the endocardial
        # surface and we append the node indices to node_numbering_list
        node_numbering_list.append(facet.entities(0))

        
# convert list to np array as it is easier to work with
triangle_node_indices = np.array(node_numbering_list)
triangle_node_indices.shape
```

#### Now plot endocardial surface

```python
fig = plt.figure(figsize=FIGSIZE)
ax = plt.axes(projection='3d')
ax.plot_trisurf(ref_coords[:,0], 
                ref_coords[:,1], 
                ref_coords[:,2], 
                triangles=triangle_node_indices,
                linewidths=0.5,
                alpha=0.5);

# adjust these values to change angle of 3D plot
ax.view_init(elev=25, azim=-179)
```

## 7. Find indices of the element that each of the endocardial surface triangle is part of

Each endocardial surface triangle is part of only **one** element in the mesh. We can find the index of this element by, for each surface triangle (``triangle_node_indices``), looping over all elements in the mesh (``mesh.cells()``), and checking if all three points in the surface triangle are in a given element (``if ((tri_i[0] in element_j) and (tri_i[1] in element_j) and (tri_i[2] in element_j))``), we have found the element we are looking for, and we append its index to the placeholder list (``elem_indices_holder``) and break from the loop before moving on to the next surface triangle.

```python
elem_indices_holder = np.zeros((triangle_node_indices.shape[0],1), dtype=np.int64)

# loop over all endocardial triangles
for i, tri_i in enumerate(triangle_node_indices):
    # loop over all elements in the mesh (slow)
    for j, element_j in enumerate(mesh.cells()):   
        # if all triangle nodes are in element_j, this is the element we are looking for and 
        # we save and break the (inner) loop
        if ((tri_i[0] in element_j) and (tri_i[1] in element_j) and (tri_i[2] in element_j)):
            elem_indices_holder[i] = j
            break
```

```python
elem_indices_holder.shape, triangle_node_indices.shape
```

The below array is useful for PINN computations

```python
endocardial_facets = np.hstack((elem_indices_holder, triangle_node_indices))
np.save(f'{save_dir}/pressure-surface-facets.npy', endocardial_facets)
```
